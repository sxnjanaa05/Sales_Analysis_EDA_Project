**Histogram.**

**BoxPlot.**

Pivot: A Pivot Table is a data summarization technique used to reorganize, group, and analyze data in a structured format.
In this project, pivot tables are used to analyze sales data based on:
- State
- Gender
- Purchase Amount

```python
New_Pivot = df.pivot_table(
    values="Amount",
    columns="Gender",
    index="State",
    aggfunc="sum"
)
```

```python
New_Pivot
```

**Heatmap**

*This heatmap represents the total sales amount generated across different states based on customer gender.*

```python
New_Pivot = df.pivot_table(
    values="Amount",
    columns="Gender",
    index="State",
    aggfunc="sum"
)
plt.figure(figsize=(15,6))

sns.heatmap(New_Pivot,annot=True,fmt=".0f", cmap="crest")

plt.title( "State Wise Amount Distribution by Gender", fontsize=16, fontweight="bold",color="#4B9C97",family="georgia")

plt.show()
```

**Heatmap**

*This heatmap displays the average purchase amount generated by male and female customers across different states.*

```python
New_Pivot = df.pivot_table(
    values="Amount",
    columns="Gender",
    index="State",
    aggfunc="mean"
)

plt.figure(figsize=(15,6))

sns.heatmap(New_Pivot,annot=True,fmt=".0f",cmap="crest")

plt.title( "State Wise Amount Distribution by Gender", fontsize=16, fontweight="bold",color="#4B9C97",family="georgia")

plt.show()
```

*This heatmap represents the maximum purchase amount recorded across different states based on customer gender.*

```python
New_Pivot = df.pivot_table(
    values="Amount",
    columns="Gender",
    index="State",
    aggfunc="max"
)

plt.figure(figsize=(15,6))

sns.heatmap(New_Pivot,annot=True,fmt=".0f",cmap="crest")

plt.title( "State Wise Amount Distribution by Gender", fontsize=16, fontweight="bold",color="#4B9C97",family="georgia")

plt.show()
```

**Heatmap**

*This heatmap represents the minimum purchase amount recorded across different states based on customer gender.*

```python
New_Pivot = df.pivot_table(
    values="Amount",
    columns="Gender",
    index="State",
    aggfunc="min"
)

plt.figure(figsize=(15,6))

sns.heatmap(New_Pivot,annot=True,fmt=".0f",cmap="crest")

plt.title( "State Wise Amount Distribution by Gender", fontsize=16, fontweight="bold",color="#4B9C97",family="georgia")

plt.show()
```
