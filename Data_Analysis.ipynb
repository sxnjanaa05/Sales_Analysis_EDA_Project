# 📊 Data Visualization 

**Data visualization is performed to represent the dataset graphically and identify patterns, trends, comparisons, and relationships using charts and graphs.**

**Count Plot**

*A count plot is used to compare the distribution of genders across different zones using the hue parameter.*

```python
plt.figure(figsize=(10,6))

Palette_1 = ["#1F5C59", "#2E7A76", "#4B9C97", "#76B7B2", "#8EC6C2"]

chart_1 = sns.countplot(x="Gender", hue="Zone", data=df,palette=Palette_1, edgecolor="Black")

plt.title("Zone Wise Gender Distribution", color="#1F5C59", fontsize=18, fontweight="bold",family="georgia")

plt.xlabel("Zone", fontsize=14, fontweight="bold")
plt.ylabel("Count", fontsize=14, fontweight="bold")

plt.xticks(rotation=45,)

plt.grid(axis="y", linestyle="--",color="#4B9C97")

for t in chart_1.containers:
    plt.bar_label(t)

plt.show()
```

**Bar Plot**

*A bar plot is used to visualize the total number of orders placed by different age groups after grouping the data using groupby().*

```python
age_orders = df.groupby("Age Group")["Orders"].sum().reset_index()

plt.figure(figsize=(10,6))

chart_2 = sns.barplot(x="Age Group", y="Orders", data=age_orders, palette=Palette_2 , edgecolor="Black")

plt.title("Age Group Wise Orders", color="#4B9C97", fontsize=18, fontweight="bold", family="georgia" )

plt.xlabel("Age Group", fontsize=14, fontweight="bold")
plt.ylabel("Total Orders", fontsize=14, fontweight="bold")

plt.xticks(rotation=45,)

plt.grid(axis="y", linestyle="--",color="#4B9C97")

for y in chart_2.containers:
    plt.bar_label(y)

plt.show()
```

**Line Plot**

*A line plot is used to visualize the top 15 sectors with the highest number of orders after grouping and sorting the data based on total orders.*

```python
Top_Sec = df.groupby("Sector",as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(15)

Top_Sec

plt.figure(figsize=(10,6))

sns.lineplot(x="Sector",y="Orders",data=Top_Sec,marker="o",color="#1F5C59")

plt.title("Top 15 Sectors Order Wise",color="#1F5C59",fontweight="bold",fontsize=16,family="georgia")

plt.xticks(rotation=30)

plt.xlabel("Age Group", fontsize=14, fontweight="bold")
plt.ylabel("Total Orders", fontsize=14, fontweight="bold")

plt.grid(axis="y", linestyle="--",color="#4B9C97")

plt.show()
```

**Pie Plot**

*This pie chart represents the total amount contributed by different age groups.*

```python
Pie = df.groupby("Age Group", as_index=False)["Amount"].sum()

plt.figure(figsize=(10,6))

Palette_2 = ["#1F5C59", "#2E7A76", "#4B9C97", "#76B7B2", "#8EC6C2",  "#A7D6D3", "#C1E5E3"]

plt.pie( Pie["Amount"], labels=Pie["Age Group"], startangle=42, autopct="%1.f%%", colors=Palette_2)

plt.title( "Age Wise Amount", color="#1F5C59", fontweight="bold", fontsize=16,family="georgia")

plt.show()
```

**Bar Plot**


*A bar plot is used to visualize the top 10 states with the highest sales amount after grouping and sorting the data based on total sales.*

```python
plt.figure(figsize=(10,6))

Palette_3 = ["#1F5C59", "#2E7A76", "#4B9C97", "#76B7B2", "#8EC6C2",  "#A7D6D3", "#C1E5E3", "#DAF1F0", "#EDF9F8", "#F6FCFC"]

State_1 = df.groupby("State", as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)

chart_3 = sns.barplot(x="State", y="Amount", data=State_1, palette=Palette_3, edgecolor="Black")

plt.title("Top 10 States by Sales Amount", fontsize=18, fontweight="bold", color="#4B9C97",family="georgia")

plt.xlabel("State", fontsize=14, fontweight="bold")
plt.ylabel("Total Amount", fontsize=14, fontweight="bold")

plt.xticks(rotation=30)
plt.yticks(fontsize=11)

plt.grid(axis="y", linestyle="--", color="#8EC6C2")

plt.show()
```

**Bar Plot**

*A bar plot is used to visualize the top 10 states with the highest orders after grouping and sorting the data based on orders.*

```python
plt.figure(figsize=(10,6))

Palette_3 = ["#1F5C59", "#2E7A76", "#4B9C97", "#76B7B2", "#8EC6C2",  "#A7D6D3", "#C1E5E3", "#DAF1F0", "#EDF9F8", "#F6FCFC"]

State_1 = df.groupby("State", as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)

chart_4 = sns.barplot(x="State", y="Orders", data=State_1, palette=Palette_3, edgecolor="Black")

plt.title("Top 10 States by Orders", fontsize=18, fontweight="bold", color="#4B9C97",family="georgia")

plt.xlabel("State", fontsize=14, fontweight="bold")
plt.ylabel("Total Amount", fontsize=14, fontweight="bold")

plt.xticks(rotation=30)
plt.yticks(fontsize=11)

plt.grid(axis="y", linestyle="--", color="#8EC6C2")

for i in chart_4.containers:
    chart_4.bar_label(i)


plt.show()
```

**Pie Plot**

*A pie chart is used to visualize the contribution of different product categories to the total sales amount.*

```python
top_category = df.groupby("Product_Category", as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(5)

plt.figure(figsize=(10,6))

plt.pie(top_category["Amount"], labels=top_category["Product_Category"], autopct="%1.1f%%", startangle=90, colors=Palette_1)

plt.title("Top Product Categories by Amount", fontsize=18, fontweight="bold", color="#4B9C97",family="georgia")

plt.show()
```

**Bar Plot**

*A bar plot is used to visualize the top 10 customers contributing the highest sales amount.*

```python
top_customer = df.groupby("Cust_name", as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)

plt.figure(figsize=(10,6))

chart_5 = sns.barplot( x="Cust_name", y="Amount", data=top_customer, palette=Palette_3, edgecolor="Black")

plt.title( "Top 10 Customers by Sales Amount", fontsize=18, fontweight="bold", color="#4B9C97",family="georgia")

plt.xlabel("Customer Name", fontsize=14, fontweight="bold")
plt.ylabel("Total Amount", fontsize=14, fontweight="bold")

plt.xticks(rotation=30)

plt.grid(axis="y", linestyle="--",color="#8EC6C2")

for i in chart_5.containers:
    chart_5.bar_label(i)

plt.show()
```

**Line Plot**

*A line plot is used to analyze the total sales amount generated across different Sectors.*

```python
sector_amount = df.groupby("Sector", as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

plt.figure(figsize=(10,6))

sns.lineplot( x="Sector", y="Amount", data=sector_amount, marker="o", linewidth=3, color="#1F5C59")

plt.title("Sector Wise Sales Amount", fontsize=18, fontweight="bold", color="#1F5C59",family="georgia")

plt.xlabel("Sector", fontsize=14, fontweight="bold")
plt.ylabel("Total Amount", fontsize=14, fontweight="bold")

plt.xticks(rotation=45)

plt.grid(axis="y", linestyle="--",color="#8EC6C2")

plt.show()
```

**Count Plot**

```python
top_category = df.groupby("Product_Category", as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)

top_10 = df[df["Product_Category"].isin(top_category["Product_Category"])]

plt.figure(figsize=(10,6))

chart_6 = sns.countplot(x="Product_Category", hue="Gender", data=top_10, palette= ["#1F5C59", "#76B7B2"], edgecolor="Black")

plt.title("Top 10 Product Category vs Gender Distribution", fontsize=18, fontweight="bold", color="#1F5C59",family="georgia")

plt.xlabel("Product Category")
plt.ylabel("Count")

plt.xticks(rotation=30)

plt.grid(axis="y", linestyle="--",color="#8EC6C2")

for i in chart_6.containers:
    chart_6.bar_label(i)

plt.show()
```
